# Limpieza y Validación de Dataframes

```python
import pandas as pd
```

```python
df = pd.read_table("../data/raw/Data_Fifa - case dirty.txt",sep=';')
```

```python
df.head(10)
```

## Tipos de Datos

```python
df.dtypes
```

## Renombras columnas

```python
df = df.rename(columns={'wage_dollars': 'wage_euros'})
```

```python
df.head()
```

## Reemplazar valores

```python
df['age']   =  df['age'].replace('31 años',31)
```

```python
df.head()
```

## Cambiar tipos de datos

```python
df.dtypes
```

```python
df['age']=df['age'].astype(float) #str, int, category
```

```python
df.dtypes
```

```python
df.head(10)
```

## Borrar duplicados

```python
df.head(10)
```

```python
df = df.drop_duplicates()
```

```python
df.head(10)
```

## Borrar fila

```python
df=df.drop([9], axis=0)
```

```python
df.head(10)
```

## Borrar columnas

elimina la columna llamada 'variable_error'. axis=1 indica eliminar columnas.

```python
df=df.drop(['variable_error'], axis=1)
```

```python
df.head(7)
```

## Completar Vacios

```python
df.loc[0,'nationality']='Argentina' # loc location

df.head(10)
```

```python
"""
 crea un nuevo DataFrame con un índice numérico secuencial (0, 1, 2, ...) 
 y asigna este nuevo DataFrame a la variable df, sobrescribiendo el DataFrame original.
"""
df=df.reset_index(drop=True)

df.head(10)
```
